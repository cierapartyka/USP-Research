In this program, plots about both USPs and Hot Jupiters will be **grey**. 
Plots about USPs only will be in **blue**.
Plots about Hot Jupiters only will be in **orange**. 

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import statistics
import math
```

```python
## USP Certain Metallicity

filename = "ajaa7b7ct1_ascii.csv"

uspdata = np.genfromtxt(filename, delimiter = ",", skip_header = 5, 
                     names = ["ID", "T_eff", "Teff_upper_unc", "Teff_lower_unc", 
                               "logg", "logg_upper_unc", "logg_lower_unc", 
                               "metal", "metal_upper_unc", "metal_lower_unc", 
                               "R_sstar", "R_sstar_upper_unc", "R_sstar_lower_unc", 
                               "M_sstar", "M_sstar_upper_unc", "M_sstar_lower_unc", 
                               "R_p", "R_p_upper_unc", "R_p_lower_unc", "P_orb"], dtype=None)
uspmetal = uspdata["metal"]


## Hot Jupiter Certain Metallicity

filename = "ajaa7b7ct2_ascii.csv"

hjdata = np.genfromtxt(filename, delimiter = ",", skip_header = 5, 
                     names = ["ID", "T_eff", "Teff_upper_unc", "Teff_lower_unc", 
                               "logg", "logg_upper_unc", "logg_lower_unc", 
                               "metal", "metal_upper_unc", "metal_lower_unc", 
                               "R_sstar", "R_sstar_upper_unc", "R_sstar_lower_unc", 
                               "M_sstar", "M_sstar_upper_unc", "M_sstar_lower_unc", 
                               "R_p", "R_p_upper_unc", "R_p_lower_unc", "P_orb"], dtype=None)
hjmetal = hjdata["metal"]
```

```python
# KS Test on Nominal USP and HJ Metalicities
# From File "CPW Recreating Winn Data". 
# Fiddled with the bin sizes so that it matches up with the figure in Josh Winn's paper.
# Blue USP 
usp_metal_graph = plt.hist(uspdata["metal"], bins = [-.59, -.49, -.39, -.29, -.19, -.09, .02, .12, .21, .32, .41], density = True, histtype = "bar", label = "USP Planets", color = "grey")

# Orange HJ
hj_metal_graph = plt.hist(hjdata["metal"], bins = [-.16, -.06, .05, .13, .24, .34, .44], density = True, histtype = "step", label = "Hot Jupiters", color = "black")

# KS-Test for USP and HJ metalicity
_, pvalue = stats.ks_2samp(uspdata["metal"], hjdata["metal"])
print("KS Test pvalue: ",pvalue)
plt.ylabel("Normalized Histogram Density")
plt.xlabel("Stellar Metallicity")
plt.title("Metallicity Distribution")
plt.legend()
```

```python
## AD Test and SW test on USP Metalicities Only
plt.hist(uspmetal, histtype = "bar", density = True, label = "USP Metalicities")
plt.xlabel("Stellar Metallicity")
plt.ylabel("Normalized Histogram Density")
plt.title("USP Metalicities Only")
plt.vlines(statistics.median(uspmetal), 0, 3, color = "red", label = "Median Value")
plt.legend()

print(stats.anderson(uspmetal, dist = 'norm')) # Comparing uspmetal's population to normal distribution
w, sw_pvalue = stats.shapiro(uspmetal)
print("\nShapiro-Wilk W stat:", w, "Shapiro Wilk P value:", sw_pvalue) # If our data was perfectly normal, W = 1. 
```

```python
# AD Test and SW Test on HJ Metalicities only
plt.hist(hjmetal, histtype = "bar", density = True, label = "HJ Metalicities", color = "orange")
plt.xlabel("Stellar Metallicity")
plt.ylabel("Normalized Histogram Density")
plt.title("HJ Metalicities Only")
plt.vlines(statistics.median(hjmetal), 0, 5, color = "red", label = "Median Value")
plt.legend()

print(stats.anderson(hjmetal, dist = 'norm')) # Comparing uspmetal's population to normal distribution
w, sw_pvalue = stats.shapiro(hjmetal)
print("\nShapiro-Wilk W stat:", w, "Shapiro Wilk P value:", sw_pvalue) # If our data was perfectly normal, W = 1. 
```

```python
def bootstrap_resample(array, lower_uncertainty_array, upper_uncertainty_array):
    
    bootstrap_result_array = []
    #rand_result_array = []
    
    for elem in range(len(array)):
        rand_result = np.random.uniform(-lower_uncertainty_array[elem], upper_uncertainty_array[elem])
        #rand_result_array.append(rand_result)
        bootstrap_data_point = array[elem] + rand_result
        bootstrap_result_array.append(bootstrap_data_point)
        np.asarray(bootstrap_result_array)  
    return np.asarray(bootstrap_result_array)
```

```python
# Bootstrapping 1000 times, and performing a 2samp KS test, an AD test on each data set, and a SW test on each data set

this_many_times = 1000

ks_list = []
usp_ad_list = []
hj_ad_list = []
usp_sw_w_list = []
hj_sw_w_list = []
usp_sw_p_list = []
hj_sw_p_list = []

for number in range(this_many_times):
    
    usp_bootstrap_result_list = []
    hj_bootstrap_result_list = []

## USP Resampling
    for data_point in uspdata["metal"]:
        rand_result = np.random.uniform(-0.04, 0.041)
        usp_bootstrap_data_point = data_point+rand_result
        usp_bootstrap_result_list.append(usp_bootstrap_data_point)

## HJ Resampling
    for data_point in hjdata["metal"]:
        rand_result = np.random.uniform(-0.04, 0.041)
        hj_bootstrap_data_point = data_point + rand_result
        hj_bootstrap_result_list.append(hj_bootstrap_data_point)

## KS Test on those lists
    _, pvalue = stats.ks_2samp(usp_bootstrap_result_list, hj_bootstrap_result_list)
    ks_list.append(pvalue)

# AD Test on those lists
    usp_statistic, _, _ = stats.anderson(usp_bootstrap_result_list)
    usp_ad_list.append(usp_statistic)
    hj_statistic, _, _ = stats.anderson(hj_bootstrap_result_list)
    hj_ad_list.append(hj_statistic)

# SW Test on those lists     
    usp_w, usp_p = stats.shapiro(usp_bootstrap_result_list)
    usp_sw_w_list.append(usp_w)
    usp_sw_p_list.append(usp_p)
    hj_w, hj_p = stats.shapiro(hj_bootstrap_result_list)
    hj_sw_w_list.append(hj_w)
    hj_sw_p_list.append(hj_p)
    
plt.hist(ks_list, bins = "auto", density = True, histtype = "bar", log = True, color = "grey")
plt.vlines(.0003, 0, 3500, colors = "cyan", label = "P = 0.0003")
plt.vlines(statistics.median(ks_list), 0, 3500, colors = "red", label = "Median Value")
plt.xlabel("P-values")
plt.ylabel("Normalized Histogram Density")
plt.title("KS Test on Bootstrapped USP and HJ")
plt.legend()
```

```python
plt.hist(usp_ad_list, density = True, label = "Bootstrapped USP AD Statistics")
plt.xlabel("AD Statistics")
plt.ylabel("Normalized Histogram Density")
plt.title("Distribution of Bootstrapped USP AD Statistics")
plt.vlines(statistics.median(usp_ad_list), 0, 4.5, color = "Red", label = "Median Value")
plt.legend()
```

```python
plt.hist(hj_ad_list, density = True, label = "Bootstrapped HJ AD Statistics", color = "orange")
plt.xlabel("AD Statistics")
plt.ylabel("Normalized Histogram Density")
plt.title("Distribution of Bootstrapped HJ AD Statistics")
plt.vlines(statistics.median(hj_ad_list), 0, 3, color = "Red", label = "Median Value")
plt.legend()
```

```python
plt.hist(usp_sw_w_list, density = True, label = "Bootstrapped USP SW W Statistics")
plt.xlabel("W Statistics")
plt.ylabel("Normalized Histogram Density")
plt.title("Distribution of Bootstrapped USP SW W Statistics")
plt.vlines(statistics.median(usp_sw_w_list), 0,50, color = "Red", label = "Median Value")
plt.legend()
```

```python
plt.hist(hj_sw_w_list, density = True, label = "Bootstrapped HJ SW W Statistics", color = "orange")
plt.xlabel("W Statistics")
plt.ylabel("Normalized Histogram Density")
plt.title("Distribution of Bootstrapped HJ SW W Statistics")
plt.vlines(statistics.median(hj_sw_w_list), 0,25, color = "Red", label = "Median Value")
plt.legend()
```

```python
plt.hist(usp_sw_p_list, density = True, label = "Bootstrapped USP SW P Statistics")
plt.xlabel("P Statistics")
plt.ylabel("Normalized Histogram Density")
plt.title("Distribution of Bootstrapped USP SW P Statistics")
plt.vlines(statistics.median(usp_sw_p_list), 0,65, color = "Red", label = "Median Value")
plt.legend()
```

```python
plt.hist(hj_sw_p_list, density = True, label = "Bootstrapped HJ SW P Statistics", color = "orange")
plt.xlabel("P Statistics")
plt.ylabel("Normalized Histogram Density")
plt.title("Distribution of Bootstrapped HJ SW P Statistics")
plt.vlines(statistics.median(hj_sw_p_list), 0,10, color = "Red", label = "Median Value")
plt.legend()
```
