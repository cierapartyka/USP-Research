```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import statistics
#from astropy.stats import bayesian_blocks
```

```python
## USP Certain Metallicity

filename = "ajaa7b7ct1_ascii.csv"

uspdata = np.genfromtxt(filename, delimiter = ",", skip_header = 5, 
                     names = ["ID", "T_eff", "Teff_upper_unc", "Teff_lower_unc", 
                               "logg", "logg_upper_unc", "logg_lower_unc", 
                               "metal", "metal_upper_unc", "metal_lower_unc", 
                               "R_sstar", "R_sstar_upper_unc", "R_sstar_lower_unc", 
                               "M_sstar", "M_sstar_upper_unc", "M_sstar_lower_unc", 
                               "R_p", "R_p_upper_unc", "R_p_lower_unc", "P_orb"], dtype=None)
uspmetal = uspdata["metal"]
print(uspmetal)
print(uspdata["metal_upper_unc"])
print(uspdata["metal_lower_unc"])
print()
print(len(uspmetal))
```

```python
## Hot Jupiter Certain Metallicity

filename = "ajaa7b7ct2_ascii.csv"

hjdata = np.genfromtxt(filename, delimiter = ",", skip_header = 5, 
                     names = ["ID", "T_eff", "Teff_upper_unc", "Teff_lower_unc", 
                               "logg", "logg_upper_unc", "logg_lower_unc", 
                               "metal", "metal_upper_unc", "metal_lower_unc", 
                               "R_sstar", "R_sstar_upper_unc", "R_sstar_lower_unc", 
                               "M_sstar", "M_sstar_upper_unc", "M_sstar_lower_unc", 
                               "R_p", "R_p_upper_unc", "R_p_lower_unc", "P_orb"], dtype=None)
hjmetal = hjdata["metal"]
print(hjmetal)
print(hjdata["metal_lower_unc"])
print()
print(len(hjmetal))
```

```python
## KS test on nominal metalicities
## Even if we do the ks test 1000 times, the output will be the same each time, and that's fine
## Expected value is 0.0003

_, pvalue = stats.ks_2samp(uspmetal, hjmetal)
print("P-Value: " + str(pvalue))
```

```python
## Bootstrap resampling function
## Upper and lower uncertainties have the exact same range for both data sets

def bootstrap_resample(array, lower_uncertainty_array, upper_uncertainty_array):
    
    bootstrap_result_array = []
    
    for elem in range(len(array)):
        rand_result = np.random.uniform(-lower_uncertainty_array[elem], upper_uncertainty_array[elem])
        bootstrap_data_point = array[elem] + rand_result
        bootstrap_result_array.append(bootstrap_data_point)
        np.asarray(bootstrap_result_array)

## Uncomment to perform a test on this function
#    original_length = len(array)
#    post_bootstrap_length = len(np.asarray(bootstrap_result_array))
#    
#    if original_length == post_bootstrap_length:
#        print("No elements were harmed in the making of this array.")
#    else: 
#        print("You hecked up a perfectly good array is what you did.")
    
    return np.asarray(bootstrap_result_array)
```

```python
## Resampling multiple times, then running a KS-Test for each of those times, resulting in an array of different p_values

bootstrap_pvalue_list = []
this_many_times = 1000

for number in range(this_many_times):
    usp_bootstrap = bootstrap_resample(uspmetal, uspdata["metal_lower_unc"], uspdata["metal_upper_unc"])
    hj_bootstrap = bootstrap_resample(hjmetal, hjdata["metal_lower_unc"], hjdata["metal_upper_unc"])
    _, pvalue = stats.ks_2samp(usp_bootstrap, hj_bootstrap)
    bootstrap_pvalue_list.append(pvalue)
```

```python
## Plotting with Statistics

def statsy_plot(array):
    
    maximum_x = max(array)
    median = statistics.median(array)
    mode = statistics.mode(array)
    
    print("Maximum = ", maximum_x)
    print("Median = ", median)
    print("Mode = ", mode)
    
    max_y = 0
    for elem in array:
        if elem == mode:
            max_y += 1
    
    plt.hist(array, bins = 100, density = True, histtype = "step", log = True)
    med_line = plt.vlines(median, 0, max_y, colors = "orange", linewidth = 1, label = "Median")
    winn_pvalue = plt.vlines(0.0003, 0, max_y, colors = "r", linewidth = 1, label = "P - 0.0003")
    max_line = plt.vlines(maximum_x, 0, max_y, colors = "black", linewidth = 1, label = "Max")
    mode_line = plt.vlines(mode, 0, max_y, colors = "g", linewidth = 1, label = "Mode")
    plt.legend()
```

```python
statsy_plot(bootstrap_pvalue_list)
plt.title("Bootstrapped Metalicities KS list")
#plt.xscale('log')
plt.ylabel("Density of P-Values")
```

```python
#!#!#!#!#!#!#!#!#!#!#!#!#!#!#!#!#!# Note the significance of this result, Maybe it's not weird that p_values /= 1. 
this_many_times = 10000

usp_bootstrap_pvalue_list = []
for i in range(this_many_times):
    usp1 = bootstrap_resample(uspmetal, uspdata["metal_lower_unc"], uspdata["metal_upper_unc"])
    usp2 = bootstrap_resample(uspmetal, uspdata["metal_lower_unc"], uspdata["metal_upper_unc"])
    _, pvalue = stats.ks_2samp(usp1, usp2)
    usp_bootstrap_pvalue_list.append(pvalue)
    
plt.hist(usp_bootstrap_pvalue_list, bins = 100, density = True, histtype = "step", log = True)
plt.title("Bootstrapped USP vs Bootstrapped USP")
plt.ylabel("Density of P-values")
print("Minimum USP P-Value: ",min(usp_bootstrap_pvalue_list))
plt.show()

hj_bootstrap_pvalue_list = []
for i in range(this_many_times):
    hj1 = bootstrap_resample(hjmetal, hjdata["metal_lower_unc"], hjdata["metal_upper_unc"])
    hj2 = bootstrap_resample(hjmetal, hjdata["metal_lower_unc"], hjdata["metal_upper_unc"])
    _, pvalue = stats.ks_2samp(hj1, hj2)
    hj_bootstrap_pvalue_list.append(pvalue)
plt.hist(hj_bootstrap_pvalue_list, bins = 100, density = True, histtype = "step", log = True)
plt.title("Bootstrapped HJ vs Bootstrapped HJ")
plt.ylabel("Density of P-values")
print("Minimum HJ P-value: ", min(hj_bootstrap_pvalue_list))
plt.show()
```

```python
## Sneaking HJs into USPs and Bootstrapping

this_many_sneaks = 4 # Can only go up to 23
this_many_times = 1000

usp_copy = uspmetal
usp_indices = np.zeros(23)
hj_indices = np.zeros(64)
index_count = 0

for i in range(this_many_sneaks):
    
    sneaky_index = np.random.randint(1,24)
    while sneaky_index in hj_indices: # Checks to make sure you're working with unique elements
        sneaky_index = np.random.randint(1,24)
    sneaky_hj = hjmetal[sneaky_index-1] # This makes a copy of a HJ element
    
    target_index = np.random.randint(1,65)
    while target_index in usp_indices:
        target_index = np.random.randint(1,65)
    usp_copy[target_index-1] = sneaky_hj # This performs the replacement
    
    hj_indices[index_count] = sneaky_index
    usp_indices[index_count] = target_index
    index_count += 1
    
    vs_usp_ks_scores = []
    hj_usp_ks_scores= []
    
    for k in range(this_many_times):
        # Bootstrap USP_copy vs Bootstrap Original USP
        bootstrap_usp_copy = bootstrap_resample(usp_copy, uspdata["metal_lower_unc"], uspdata["metal_upper_unc"])
        bootstrap_usp = bootstrap_resample(uspmetal, uspdata["metal_lower_unc"], uspdata["metal_upper_unc"])
        _, usp_pvalue = stats.ks_2samp(bootstrap_usp, bootstrap_usp_copy)
        vs_usp_ks_scores.append(usp_pvalue)
        
        # Bootstrap HJ vs Bootstrap Original USP
        bootstrap_hj = bootstrap_resample(hjmetal, hjdata["metal_lower_unc"], hjdata["metal_upper_unc"])
        _, hj_usp_pvalue = stats.ks_2samp(bootstrap_usp_copy, bootstrap_hj)
        hj_usp_ks_scores.append(hj_usp_pvalue)
    
    # Plot a graph after each time you swapped     
    vs_usp_ks_scores_graph = statsy_plot(vs_usp_ks_scores)
    plt.title("Sneaky USP vs Original USP. {} replaced".format(this_many_sneaks))
    plt.ylabel("Density of P-values")
    plt.show()
    
    hj_usp_ks_scores_graph = statsy_plot(hj_usp_ks_scores)
    plt.title("Sneaky USP vs Original HJ. {} replaced".format(this_many_sneaks))
    plt.ylabel("Density of P-values")
    plt.show()
```
